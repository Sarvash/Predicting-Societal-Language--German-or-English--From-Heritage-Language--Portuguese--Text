```python
import numpy as np
from sklearn.impute import SimpleImputer

# The values of these columns did not make sense as a result of an error in CTAP's calculations.
columns_to_remove = [
    'Lexical Sophistication Feature: Age of Acquisition (AW Type)',
    'Lexical Sophistication Feature: Concreteness (FW Token)'
]
df = df.drop(columns=columns_to_remove, errors='ignore')

feature_columns = [col for col in df.columns if col not in ['Text_Title', 'Country', 'Proficiency', 'Topic', 'General Topic']]

# Impute missing values with mean; these are possibly as a result of issues with CTAP
imputer = SimpleImputer(strategy='mean')
df[feature_columns] = imputer.fit_transform(df[feature_columns])
```

# Clustering-Based Undersampling

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import pairwise_distances_argmin_min

# Get unique proficiency levels
proficiency_levels = df['Proficiency'].unique()
balanced_data = []

for prof in proficiency_levels:
    # Subset data for the current proficiency level
    df_prof = df[df['Proficiency'] == prof]
    
    # Separate DE and UK samples
    DE_samples = df_prof[df_prof['Country'] == 'DE'].copy()
    UK_samples = df_prof[df_prof['Country'] == 'UK'].copy()
    
    # Determine which group is larger
    num_DE = len(DE_samples)
    num_UK = len(UK_samples)
    
    if num_DE > num_UK:
        # DE is larger; downsample DE
        larger_group = DE_samples
        smaller_group = UK_samples
    else:
        # UK is larger; downsample UK
        larger_group = UK_samples
        smaller_group = DE_samples
    
    # Number of samples to match
    n_samples = len(smaller_group)
    
    # Features for the larger group
    X_larger = larger_group[feature_columns].values
    
    # Perform PCA to reduce dimensions
    pca = PCA(n_components=10, random_state=42)
    X_larger_pca = pca.fit_transform(X_larger)
    
    # Apply KMeans clustering to the larger group
    kmeans = KMeans(n_clusters=n_samples, random_state=42)
    kmeans.fit(X_larger_pca)
    larger_group['cluster'] = kmeans.labels_
    
    # Find the sample closest to each cluster centroid
    centroids = kmeans.cluster_centers_
    closest, _ = pairwise_distances_argmin_min(centroids, X_larger_pca)
    selected_samples = larger_group.iloc[closest]
    
    # Combine the downsampled larger group with the smaller group
    df_balanced_level = pd.concat([selected_samples, smaller_group])
    balanced_data.append(df_balanced_level)

# Concatenate balanced data from all proficiency levels
df_balanced = pd.concat(balanced_data)

df_balanced.reset_index(drop=True, inplace=True)
print(df_balanced['Country'].value_counts())
```

# Training & Testing

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix
from interpret.glassbox import ExplainableBoostingClassifier
from interpret import show
import matplotlib.pyplot as plt
import seaborn as sns

# Encode 'Country' as binary labels ('DE' -> 0, 'UK' -> 1)
country_mapping = {'DE': 0, 'UK': 1}
df_balanced['Country'] = df_balanced['Country'].map(country_mapping)

# Define features and target variable
feature_columns = [col for col in df_balanced.columns if col not in ['Text_Title', 'Country', 'Topic', 'General Topic']]
X = df_balanced[feature_columns]
y = df_balanced['Country']

feature_names = X.columns.tolist()

# For Proficiency, treat it as categorical. Others are continuous.
feature_types = []
for feature in feature_names:
    if feature == 'Proficiency':
        feature_types.append(['B1', 'B2', 'C1'])
    else:
        feature_types.append('continuous')

# Create list of interactions between 'Proficiency' and other features
proficiency_index = feature_names.index('Proficiency')
interaction_indices = [(proficiency_index, i) for i in range(len(feature_names)) if i != proficiency_index]

# We'll do a manual cross-validation to get out-of-fold predictions
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

y_true_all = []
y_pred_all = []
X_indices_all = []

fold_accuracies = []
fold_f1_scores = []
fold_precisions = [] 
fold_recalls = [] 

for train_idx, test_idx in cv.split(X, y):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    ebm_fold = ExplainableBoostingClassifier(
        feature_names=feature_names,
        feature_types=feature_types,
        interactions=interaction_indices,
        random_state=42,
        exclude="mains"
    )
    ebm_fold.fit(X_train, y_train)

    y_pred_fold = ebm_fold.predict(X_test)

    y_true_all.append(y_test)
    y_pred_all.append(y_pred_fold)
    X_indices_all.append(test_idx)

    # Compute fold metrics
    fold_acc = accuracy_score(y_test, y_pred_fold)
    fold_precision = precision_score(y_test, y_pred_fold)
    fold_recall = recall_score(y_test, y_pred_fold)
    fold_f1 = f1_score(y_test, y_pred_fold)
    
    fold_accuracies.append(fold_acc)
    fold_precisions.append(fold_precision)
    fold_recalls.append(fold_recall)
    fold_f1_scores.append(fold_f1)

y_true_all = pd.concat(y_true_all)
y_pred_all = np.concatenate(y_pred_all)
X_indices_all = np.concatenate(X_indices_all)

# Sort predictions according to original indexing
order = np.argsort(X_indices_all)
y_true_all = y_true_all.iloc[order]
y_pred_all = y_pred_all[order]

# Evaluate cross-validation performance
acc_mean = np.mean(fold_accuracies)
acc_std = np.std(fold_accuracies, ddof=1)
prec_mean = np.mean(fold_precisions)
prec_std = np.std(fold_precisions, ddof=1)
rec_mean = np.mean(fold_recalls)
rec_std = np.std(fold_recalls, ddof=1)
f1_mean = np.mean(fold_f1_scores)
f1_std = np.std(fold_f1_scores, ddof=1)

print("Cross-Validation Results (Out-of-Fold Predictions):")
print(f"Accuracy: {acc_mean:.2f} ± {acc_std:.2f}")
print(f"Precision: {prec_mean:.2f} ± {prec_std:.2f}")
print(f"Recall: {rec_mean:.2f} ± {rec_std:.2f}")
print(f"F1 Score: {f1_mean:.2f} ± {f1_std:.2f}")

######################################################
# Performance by Proficiency Level using OOF predictions
######################################################
proficiency_levels = X['Proficiency'].unique()

# Align X, y, and proficiency with the out-of-fold order
X_oof = X.iloc[X_indices_all].iloc[order]
y_oof = y_true_all

class_names = ["German", "English"]  # 0 -> German, 1 -> English

fig, axes = plt.subplots(1, len(proficiency_levels), figsize=(6 * len(proficiency_levels), 5))
if len(proficiency_levels) == 1:
    axes = [axes]

for i, level in enumerate(proficiency_levels):
    # Filter samples by proficiency
    mask = (X_oof['Proficiency'] == level)
    y_true_subset = y_oof[mask]
    y_pred_subset = y_pred_all[mask]

    # Compute confusion matrix for this subset
    cm = confusion_matrix(y_true_subset, y_pred_subset)

    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[i],
                xticklabels=class_names, yticklabels=class_names)
    axes[i].set_title(f'Proficiency: {level}')
    axes[i].set_xlabel('Predicted Majority Language')
    axes[i].set_ylabel('Majority Language')

plt.tight_layout()
plt.show()

print("\nPerformance by Proficiency Level (Out-of-Fold):")
for level in proficiency_levels:
    mask = (X_oof['Proficiency'] == level)
    y_true_subset = y_oof[mask]
    y_pred_subset = y_pred_all[mask]
    acc_level = accuracy_score(y_true_subset, y_pred_subset)
    precision_level = precision_score(y_true_subset, y_pred_subset)
    recall_level = recall_score(y_true_subset, y_pred_subset)
    f1_level = f1_score(y_true_subset, y_pred_subset)
    print(f"Proficiency: {level} - Accuracy: {acc_level:.2f}, Precision: {precision_level:.2f}, Recall: {recall_level:.2f}, F1: {f1_level:.2f}")

######################################################
# Retrain on full data for feature importance
######################################################
ebm_final = ExplainableBoostingClassifier(
    feature_names=feature_names,
    feature_types=feature_types,
    interactions=interaction_indices,
    random_state=42,
    exclude="mains"
)
ebm_final.fit(X, y)

ebm_global = ebm_final.explain_global()
show(ebm_global)
```

```python
print("Cross-Validation Results (Out-of-Fold Predictions):")
print(f"Accuracy: {acc_mean:.4f} ± {acc_std:.4f}")
print(f"Precision: {prec_mean:.4f} ± {prec_std:.4f}")
print(f"Recall: {rec_mean:.4f} ± {rec_std:.4f}")
print(f"F1 Score: {f1_mean:.4f} ± {f1_std:.4f}")

print("\nPerformance by Proficiency Level (Out-of-Fold):")
for level in proficiency_levels:
    mask = (X_oof['Proficiency'] == level)
    y_true_subset = y_oof[mask]
    y_pred_subset = y_pred_all[mask]
    acc_level = accuracy_score(y_true_subset, y_pred_subset)
    precision_level = precision_score(y_true_subset, y_pred_subset)
    recall_level = recall_score(y_true_subset, y_pred_subset)
    f1_level = f1_score(y_true_subset, y_pred_subset)
    print(f"Proficiency: {level} - Accuracy: {acc_level:.4f}, "
          f"Precision: {precision_level:.4f}, Recall: {recall_level:.4f}, F1: {f1_level:.4f}")
```

# Feature Importance Ranking

```python
# Get the feature importance data
feature_importances = ebm_global.data()

# Extract features and their importance scores
features = feature_importances['names']
importances = feature_importances['scores']

# Combine them into a sorted list of tuples (importance, feature)
sorted_features = sorted(zip(importances, features), reverse=True)

print("Top 30 Most Important Features:")
for rank, (importance, feature) in enumerate(sorted_features[:30], start=1):
    print(f"{rank}. {feature}: {importance:.4f}")
```

```python
import pandas as pd
import numpy as np

def analyze_category_importance(feature_importances, feature_names, categorization_file):
    """
    Analyze the average importance and percentage contribution of features by category.
    """
    categories_df = pd.read_csv(categorization_file)
    
    # Create a dictionary of importance scores
    # Remove 'Proficiency & ' prefix from feature names to match with CSV
    importance_dict = {
        name.replace('Proficiency & ', ''): imp 
        for imp, name in zip(feature_importances, feature_names)
    }
    
    # Add importance scores to the categories DataFrame
    categories_df['Importance'] = categories_df['Feature'].map(importance_dict)
    
    # Calculate total importance across all features
    total_importance = categories_df['Importance'].sum()
    
    # Calculate metrics by category
    category_importance = categories_df.groupby('Category').agg({
        'Importance': [
            ('Mean_Importance', 'mean'),
            ('Sum_Importance', 'sum'),
            ('Feature_Count', 'count'),
            ('Std_Importance', 'std')
        ]
    }).reset_index()
    
    # Flatten column names
    category_importance.columns = ['Category', 'Mean_Importance', 'Sum_Importance', 
                                 'Feature_Count', 'Std_Importance']
    
    # Calculate percentage metrics
    category_importance['Total_Importance_Percentage'] = (
        category_importance['Sum_Importance'] / total_importance * 100
    )
    
    category_importance['Average_Importance_Percentage'] = (
        category_importance['Mean_Importance'] / category_importance['Mean_Importance'].sum() * 100
    )
    
    category_importance = category_importance.sort_values('Mean_Importance', ascending=False)
    
    return category_importance

importances = [imp for imp, _ in sorted_features]
features = [feat for _, feat in sorted_features]
results = analyze_category_importance(
    feature_importances=importances,
    feature_names=features,
    categorization_file='Feature Categorization.csv'
)

# Print results
print("\nCategory Importance Analysis:")
print("-----------------------------")
for _, row in results.iterrows():
    print(f"\nCategory: {row['Category']}")
    print(f"Mean Importance: {row['Mean_Importance']:.4f}")
    print(f"Total Importance %: {row['Total_Importance_Percentage']:.2f}%")
    print(f"Average Importance %: {row['Average_Importance_Percentage']:.2f}%")
    print(f"Standard Deviation: {row['Std_Importance']:.4f}")
    print(f"Number of Features: {row['Feature_Count']}")
```
